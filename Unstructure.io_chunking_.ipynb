```python
# Function to format the output
def format_elements(elements):
    formatted_output = []
    for element in elements:
        element_type = type(element).__name__
        content = getattr(element, 'text', '').strip()
        formatted_output.append(f"{element_type}: {content}")
    return formatted_output

# Get the formatted output
formatted_output = format_elements(elements)

# Print the formatted output
for line in formatted_output:
    print(line)
```

```python
import json
from unstructured.documents.elements import Element, ElementMetadata

# Custom function to convert an element to a dictionary, capturing all attributes
def element_to_dict(element):
    element_dict = {}
    for key, value in element.__dict__.items():
        if not key.startswith('_'):
            if isinstance(value, ElementMetadata):
                element_dict[key] = value.__dict__
            else:
                element_dict[key] = value
    # Include the type of the element explicitly
    element_dict["type"] = element.__class__.__name__
    return element_dict

# Function to print element details
def print_element_details(element):
    element_dict = element_to_dict(element)
    for key, value in element_dict.items():
        print(f"{key}: {value}")
    print("\n")

# Iterate through elements and print details
for element in elements:
    print_element_details(element)
```
